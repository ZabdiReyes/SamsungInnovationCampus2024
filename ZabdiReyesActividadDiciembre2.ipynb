```python
from sklearn.datasets import load_iris
data = load_iris()
data.keys()
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

iris = pd.DataFrame(data = data.data, columns=data.feature_names)
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris.corr()
```

```python
#Visualizacion de estadisticas basicas

def boxplot_iris(feature_names,dataset):
    i = 1
    plt.figure(figsize=(11,9))
    for col in feature_names:
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both', left = True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom = False)
        dataset[col].plot(kind="box",subplots = True, sharex = False)
        plt.title(col)
        i += 1
    plt.show()
boxplot_iris(iris.columns[:],iris)
```

```python


def histrogram_iris(feature_names,dataset):
    i = 1
    plt.figure(figsize=(11,9))
    for col in feature_names:
        plt.subplot(2,2,i)
        plt.axis('on')
        plt.tick_params(axis='both', left = True,
                        top=False, right=False,
                        bottom=True, labelleft=False,
                        labeltop=False, labelright=False,
                        labelbottom = False)
        dataset[col].hist()
        plt.title(col)
        i += 1
    plt.show()

histrogram_iris(iris.columns[:],iris)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x = iris
y = data.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33,random_state=42)

model = DecisionTreeClassifier(
    criterion='gini',
    splitter='best',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    random_state=None,
    max_leaf_nodes=None,
    class_weight=None

)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
