```python
#Crear un dataframe de 100*100 con valores aleatorios
import pandas as pd
import numpy as np
import random
df = pd.DataFrame(np.random.randint(0,100,size=(100, 100)))
df.head()
```

```python
#Poner titulos tanto a las filas como a las columnas fila_1,fila_2,fila_3... etc.
#Defino filas y columnas como el resultado de la dimensionalidad del dataframe haciendo uso de que df.shape = filas X columnas
filas, columnas = df.shape
df.columns = ['columna_' + str(i) for i in range(1, columnas+1)]
df.index = ['fila_' + str(i) for i in range(1, filas+1)]
df.head()
```

```python
#Aleatoriamente eliminar 5 datos de toda la data / hacerlos nan

for i in range(5):
  i = random.randint(0,100)
  j = random.randint(0,100)
  df.iloc[i,j] = np.nan
df.head()
```

```python
#imprimir las filas y columnas donde un valor sea nan
for i in range(filas):
    for j in range(columnas):
        if pd.isna(df.iloc[i,j]):
            print('Fila_', i+1, 'Columna_', j)
```

```python
#Eliminar todas las filas con valores nulos
df = df.dropna()
df.shape
```

```python
#agrupar los datos por la columna columna_1 y sacar la media
grouped_df = df.groupby('columna_1').mean()
grouped_df.head()
```

```python
#Crear una nueva columna que sea producto de la columna 3 y columna 4
df.insert(4, 'columna_3_x_4', df['columna_3'] * df['columna_4'])
df.head()
```

```python
#Filtra las filas donde la nueva columna (columna_3_x_4) >300
filtered_df = df[df['columna_3_x_4'] > 300]
filtered_df
```

```python
#normaliza las columnas del dataframe filtered_df
normalized_df=(filtered_df-filtered_df.min())/(filtered_df.max()-filtered_df.min())
normalized_df
```

```python
# Usa un pivote para reorganizar y aplica 5 funciones agregadas
pivot_table = pd.pivot_table(df, values='columna_4', index='columna_3', aggfunc=['mean', 'sum', 'median', 'min', 'max'])
pivot_table
```

```python
# Selecciona las primeras 20 filas y 20 columnas para obtener una matriz 20x20 y ese resultado has que sea igual 5 en todos sus valores
df.iloc[:20, :20] = 5
df
```
